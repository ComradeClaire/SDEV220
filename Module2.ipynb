```python
secret = 5
guess = int(input("Enter an interget between 1 and 10: "))

if guess > secret:
    print("Too High!")
elif guess < secret:
    print("Too Low!")
else:
    print("You Got It!")
```

```python
small = True
green = True

choice = str(input("Enter 'Cherry', 'Pea', 'Watermelon', or 'Pumpkin'."))

if choice == "Cherry":
    small = True
    green = False
elif choice == "Pea":
    small = True
    green = True
elif choice == "Watermelon":
    small = False
    green = True
else:
    small = False
    green = False
```

```python
guess_me = 7
number = 1

while number <= guess_me:
    print(number)
    if number < guess_me:
        print("Too Low!")
    elif number == guess_me:
        print("Found it!")
        break
    else:
        print("Oops!")
        break
    number += 1
```

```python
guess_me = 5

for number in range(10):
    print(number)
    if number < guess_me:
        print("Too Low!")
    elif number == guess_me:
        print("Found it!")
        break
    else:
        print("Oops!")
        break
```
